```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer()

# Create a dataframe
cancer_data.feature_names

df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
# df['target'] = cancer_data.target_names[cancer_data['target']]
df['target'] = cancer_data.target
df.head()
```

```python
# import train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer_data.data, cancer_data.target, test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# kernel function helps to project data into a higher dimensional space
# linear kernel is the simplest kernel function
model = SVC(kernel='linear')

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
accuracy
```
