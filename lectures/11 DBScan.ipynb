```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
dataset = np.array([
            [1,2], 
            [1, 4], 
            [1, 0], 
            [10, 2], 
            [10, 4], 
            [10, 0], 
            [5, 5], 
            [5, 6], 
            [5, 4], 
            [12, 5]
        ])
```

```python
# Define the DBScan parameters
epsilon = 3.5
min_points = 2 

# Initialize the DBSCAN Algorithm
dbscan = DBSCAN(eps = epsilon, min_samples=min_points)

labels = dbscan.fit_predict(dataset)
```

```python
labels
```

```python
# Plot the results

plt.figure(figsize=(5, 5))
plt.scatter(dataset[:, 0], dataset[:, 1], c=labels, cmap='coolwarm')
plt.title('Clusters created by DBSCAN')
plt.show()
```
